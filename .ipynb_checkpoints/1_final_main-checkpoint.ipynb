```python
import tkinter as tk
from tkinter import ttk
from PIL import Image, ImageTk

# Initialize the main window
root = tk.Tk()
root.title("Controlling Laptop using Hand Gesture")
root.geometry("900x600")  # Width x Height

# Load the background image
bg_image_path = "bk.jpg"
bg_image = Image.open(bg_image_path)
bg_image = bg_image.resize((900, 600), Image.ANTIALIAS)
bg_photo = ImageTk.PhotoImage(bg_image)

# Create a Canvas widget and set the background image
canvas = tk.Canvas(root, width=900, height=600)
canvas.pack(fill="both", expand=True)
canvas.create_image(0, 0, image=bg_photo, anchor="nw")

# Function to load an image
def load_image(path, size):
    img = Image.open(path)
    img = img.resize(size, Image.ANTIALIAS)
    return ImageTk.PhotoImage(img)
import os
import time
mouse_task='mouse_task.txt'
music_task='music_task.txt'
last_modified = 0
def update_mouse_label_text():
    
    with open(mouse_task, 'r') as file:
        content = file.read()
        btn1.text=content
def check_mouse_task_change():
    global mouse_task,last_modified
    try:
        modified_time = os.path.getmtime(mouse_task)
        if modified_time != last_modified:
            last_modified = modified_time
            update_mouse_label_text()
    except FileNotFoundError:
        btn1.set("File not found")

    root.after(1000, check_mouse_task_change)
def show_message(message):
    # Clear the window
    for widget in root.winfo_children():
        widget.destroy()
    
    # Display the message
    message_label = tk.Label(root, text=message, font=("Helvetica", 24))
    message_label.pack(expand=True)
def button1_clicked():
    print("Mouse")
    f=open('task.txt','w')
    f.write('1')
    f.close()

def button2_clicked():
    print("Music")
    f=open('task.txt','w')
    f.write('2')
    f.close()
    '''
    root.destroy()
    new_window = tk.Tk()
    new_window.title("New Window")
    new_window.geometry("1200x800")
    
    # Add a label to the new window
    label = tk.Label(new_window, text="Music Player")
    label.pack(pady=20)
    
    # Run the new window's event loop
    new_window.mainloop()
    
    '''
    

def button3_clicked():
    print("Presentation")
    f=open('task.txt','w')
    f.write('3')
    f.close()
# Load your images

image1 = load_image("mouse.png", (200, 200))
image2 = load_image("music.jpg", (200, 200))
image3 = load_image("presentation.jpg", (200, 200))
# Create three frames for the three vertical parts
frame1 = tk.Frame(root, width=300, height=600, relief="solid", borderwidth=1,bg='pink')
frame2 = tk.Frame(root, width=300, height=600, relief="solid", borderwidth=1,bg='pink')
frame3 = tk.Frame(root, width=300, height=600, relief="solid", borderwidth=1,bg='pink')

frame1.place(x=10, y=10, width=280, height=580)
frame2.place(x=310, y=10, width=280, height=580)
frame3.place(x=610, y=10, width=280, height=580)

# Place widgets in the first frame
label1_img = ttk.Label(frame1, image=image1)
label1_img.pack(pady=(10, 10))

button1 = tk.Button(frame1, text="Mouse", command=button1_clicked,font=("Helvetica", 16),bg='pink')
button1.pack(pady=(10, 10))

label1_text = tk.Label(frame1, text="Mouse Status",bg='pink')
label1_text.pack(pady=(10, 10))

# Place widgets in the second frame
label2_img = ttk.Label(frame2, image=image2)
label2_img.pack(pady=(10, 10))

button2 = tk.Button(frame2, text="Music",command=button2_clicked,font=("Helvetica", 16),bg='pink')
button2.pack(pady=(10, 10))

label2_text = tk.Label(frame2, text="Music Status",bg='pink')
label2_text.pack(pady=(10, 10))

# Place widgets in the third frame
label3_img = ttk.Label(frame3, image=image3)
label3_img.pack(pady=(10, 10))

button3 = tk.Button(frame3, text="Presentation", command=button3_clicked,font=("Helvetica", 16),bg='pink')
button3.pack(pady=(10, 10))

label3_text = tk.Label(frame3, text="Presentation Status" ,bg='pink')
label3_text.pack(pady=(10, 10))

# Start the main event loop
root.mainloop()
```
