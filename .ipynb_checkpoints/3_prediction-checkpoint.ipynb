```python
import os
import cv2
import time
import numpy as np
from keras.models import load_model
import pyautogui
import time
import warnings
warnings.filterwarnings(action = 'ignore')


model = load_model('CNN_model.h5')

gestures = {
    1:'click',
    2:'dclick',
    3:'down',
    4:'left',
    5:'none',
    6:'right',
    7:'up'
    

}

def predict(gesture):
    img = cv2.resize(gesture, (50,50))
    img = img.reshape(1,50,50,1)
    img = img/255.0
    prd = model.predict(img)
    index = prd.argmax()
    return gestures[index]
```

```python
import cv2
old_text = ''
pred_text = ''
count_frames = 0
total_str = ''
flag = True
speed = 0.5
status='1'
def position_bottom_left(window_name, window_width, window_height):
    # Get screen width and height
    screen_width = cv2.getWindowImageRect(window_name)[2]
    screen_height = cv2.getWindowImageRect(window_name)[3]

    # Calculate the position for bottom-left corner
    x = 0
    y = screen_height - window_height+700

    # Move the window to the bottom-left corner
    cv2.moveWindow(window_name, x, y)

def create_camera_window():
    global count_frames,flag,total_str
    # Open the default camera
    cap = cv2.VideoCapture(0)

    # Check if the camera opened successfully
    if not cap.isOpened():
        print("Error: Could not open camera.")
        return

    # Create a named window
    window_name = 'Camera Feed'
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

    # Set the desired window size
    window_width = 250
    window_height = 250
    cv2.resizeWindow(window_name, window_width, window_height)

    # Position the window at the bottom-left corner
    position_bottom_left(window_name, window_width, window_height)

    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()
        frame = cv2.flip(frame, 1)
        frame = cv2.resize( frame, (250,250) )
        
        cv2.rectangle(frame, (250,250), (0,0), (0,255,0), 2)
        
        
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        thresh = cv2.threshold(grey,210,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]
      
        
        
        if count_frames > 20 and pred_text != "":
            total_str += pred_text
            count_frames = 0
        if flag == True and status=='1':
            
            pred_text = predict(thresh)
            print(pred_text)
            # If frame is read correctly ret is True
            if not ret:
                print("Error: Failed to capture image.")
                break

            # Display the resulting frame

            #f = open('output.txt','w')
            #f.write(str(pred_text))
            #f.close()

            if str(pred_text)=='click':
                res = 'click'
                #print(res)



            elif str(pred_text)=='rclick':
                res = 'rclick'
                #print(res)



            elif str(pred_text)=='dclick':
                res = 'dclick'
                #print(res) 


            elif str(pred_text)=='down':
                res = 'down'
                #print(res)


            elif str(pred_text)=='left':
                res = 'left'
                #print(res)
                f = open('presentation_input.txt','w')
                f.write('left')
                f.close()


            elif str(pred_text)=='none':
                res = 'none'
                #print(res)

            elif str(pred_text)=='right':
                res = 'right'
                #print(res)
                f = open('presentation_input.txt','w')
                f.write('right')
                f.close()




            elif str(pred_text)=='up':
                res = 'up'
                #print(res)
            cv2.putText(frame, res, (10, 20), cv2.FONT_HERSHEY_TRIPLEX, 1, (0, 0, 0))
            cv2.imshow(window_name, frame)
            old_text = pred_text



            #cv2.namedWindow(res, cv2.WINDOW_NORMAL)
            #cv2.putText(blackboard, res, (30, 70), cv2.FONT_HERSHEY_TRIPLEX, 1, (255, 0, 255))

            f = open('result.txt','w')
            f.write(str(res))
            f.close()
            if old_text == pred_text:
                count_frames += 1                
            else:
                count_frames = 0
            pred_text=''
            flag = True
        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        #rval, frame = cap.read()
        #flag=False
        
        #print('flag',flag)
        #if keypress == ord('c'):
    
        keypress = cv2.waitKey(1)

        if keypress == ord('p'):
            print('prev')
            f = open('presentation_input.txt','w')
            f.write('prev')
            f.close()
            f = open('presentation_task.txt','w')
            f.write('Previous Slide')
            f.close()
        if keypress == ord('n'):
            print('next')
            f = open('presentation_input.txt','w')
            f.write('next')
            f.close()
            f = open('presentation_task.txt','w')
            f.write('Next Slide')
            f.close()
    

    

    cv2.waitKey(1)
    #cv2.overrideredirect(True)  # Remove title bar
    # Release the capture
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    create_camera_window()
```

```python
import cv2

def position_bottom_left(window_name, window_width, window_height):
    # Get screen dimensions
    screen_width = cv2.getWindowImageRect(window_name)[2]
    screen_height = cv2.getWindowImageRect(window_name)[3]

    # Calculate the position for bottom-left corner
    x = 0
    y = screen_height - window_height+700

    # Move the window to the bottom-left corner
    cv2.moveWindow(window_name, x, y)

def create_window():
    # Create a named window
    window_name = 'Hand Gesture'
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

    # Set the desired window size
    window_width = 250
    window_height = 250
    cv2.resizeWindow(window_name, window_width, window_height)

    # Position the window at the bottom-left corner
    position_bottom_left(window_name, window_width, window_height)

    # Create a blank image to display
    image = 255 * np.ones(shape=[window_height, window_width, 3], dtype=np.uint8)

    # Display the image in the window
    cv2.imshow(window_name, image)

    # Wait until a key is pressed
    cv2.waitKey(0)

    # Destroy all windows
    cv2.destroyAllWindows()

if __name__ == "__main__":
    import numpy as np
    create_window()
```

```python
import cv2

def position_bottom_left(window_name, window_width, window_height):
    # Get screen dimensions
    screen_width = cv2.getWindowImageRect(window_name)[2]
    screen_height = cv2.getWindowImageRect(window_name)[3]

    # Calculate the position for bottom-left corner
    x = 0
    y = screen_height - window_height+500

    # Move the window to the bottom-left corner
    cv2.moveWindow(window_name, x, y)

def create_window():
    # Create a named window
    window_name = 'Bottom Left Window'
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

    # Set the desired window size
    window_width = 640
    window_height = 480
    cv2.resizeWindow(window_name, window_width, window_height)

    # Position the window at the bottom-left corner
    position_bottom_left(window_name, window_width, window_height)

    # Create a blank image to display
    image = 255 * np.ones(shape=[window_height, window_width, 3], dtype=np.uint8)

    # Display the image in the window
    cv2.imshow(window_name, image)

    # Wait until a key is pressed
    cv2.waitKey(0)

    # Destroy all windows
    cv2.destroyAllWindows()

if __name__ == "__main__":
    import numpy as np
    create_window()
```

```python
import tkinter as tk

# Function to position the window at the bottom-left corner
def position_bottom_left(root, window_width, window_height):
    screen_width = root.winfo_screenwidth()
    screen_height = root.winfo_screenheight()
    
    # Position the window at the bottom-left corner
    x = 0
    y = screen_height - window_height-100
    
    # Set the geometry of the window
    root.geometry(f'{window_width}x{window_height}+{x}+{y}')

# Main function to create and position the window
def create_window():
    root = tk.Tk()
    
    # Set the window size
    window_width = 250
    window_height = 250
    
    # Call the function to position the window
    position_bottom_left(root, window_width, window_height)
    
    # Title of the window
    root.title("Bottom Left Window")
    
    # Run the Tkinter event loop
    root.mainloop()

# Run the main function
if __name__ == "__main__":
    create_window()
```
